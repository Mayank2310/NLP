## Stemming

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

## PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"------>"+stemming.stem(word))
```

```python
stemming.stem('congrulations') ## This is the problem with stemming sometimes the words doesnt make any sense.
```

```python
stemming.stem('going')
```

### RegexpStemmer class

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min= 4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer=RegexpStemmer('ing|s$|e$|able$', min= 4)  ## We removed $ 'ing|'
```

```python
reg_stemmer.stem('ingeating')
```

```python
reg_stemmer.stem('playing')
```

```python
reg_stemmer.stem('history')
```

```python
reg_stemmer.stem('snowing')
```

### Snowball Stemmer

 It is a stemming algorithm which is also known as the Porter2 stemming algorithm as it is a better version of the Porter Stemmer since some issues of it were fixed in this stemmer.

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
for word in words:
    print(word+"------>"+snowballstemmer.stem(word))
```

```python
## lets see some examples where the stemming fails to work
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowballstemmer.stem("fairly"),snowballstemmer.stem("sportingly") ## example of snowballstemmer
```

```python
snowballstemmer.stem('goes')
```
