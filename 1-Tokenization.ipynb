```python
corpus=""" Hello I'm Mayank and Welcome to the Tokeniastion Example.
It is important to learn tokenisation for NLP and make sure to install the packages properly.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
## Tokenization 
## Sentence--> paragraph

from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents=sent_tokenize(corpus)
```

```python
print(documents)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization 
## convert paragraph --> words
## sentence --> words
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus) ## All words are splitted
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus) ## Fullstop is considered as one but not the last one.
```

```python
from nltk.tokenize import TreebankWordDetokenizer
```

```python
tokenizer=TreebankWordDetokenizer()
```

```python
tokenizer.tokenize(corpus)
```
